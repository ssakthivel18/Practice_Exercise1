# Experiment 1 Analysis (2 stimuli)

```python
#Import all necessary libraries
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Load data from Experiment 1
data1 = pd.read_csv('reactiontime.csv')
print(data1)
```

```python
# Scatterplot of data
plt.figure(figsize=(10, 5))
plt.plot(data1['trial_number'], data1['reaction_time'], 'o')

#Calculate standard deviation
std_devs1 = np.std(data1['reaction_time'])  

#Plot the error bars
plt.errorbar(data1['trial_number'], data1['reaction_time'], yerr=std_devs1, fmt='o')

#obtain m (slope) and b(intercept) of linear regression line
m, b = np.polyfit(data1['trial_number'], data1['reaction_time'], 1)

#add linear regression line to scatterplot 
plt.plot(data1['trial_number'], m*(data1['trial_number'])+b)

plt.xlabel('Trial Number')
plt.ylabel('Reaction Time (ms)')
plt.title('Reaction Time Experiment 1')
plt.legend(['Regression Line: y = {:.2f}x + {:.2f}'.format(m, b)])
plt.show()
```

```python
#plot showing accuracy through the experiment
plt.figure(figsize=(10, 5))
plt.plot(data1['trial_number'], data1['accuracy'])
plt.xlabel('Trial Number')
plt.ylabel('Accuracy')
plt.title('Reaction Time Experiment 1')
plt.show()
```

```python
#Summary data from the experiment
print('The average reaction time for this experiment was', data1['reaction_time'].mean(), 'ms.')
print('The standard deviation for this experiment was', std_devs1, 'ms.')
print('The average accuracy for this experiment was', data1['accuracy'].mean(), "%.")
```
